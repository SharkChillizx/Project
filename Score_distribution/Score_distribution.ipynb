```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

<h2>Function</h2>

```python
def comparision_visual_hist(original_score_list, final_score_list):
    ax1 = sns.distplot(original_score_list, kde_kws={'label': 'Association Score'})
    ax2 = sns.distplot(final_score_list, color='red', kde_kws={'label': 'Calibrated Risk Score'})
    ax1.set_xlim(0.0,1.0)
    ax2.set_xlim(0.0,1.0)
    ax1.set(yticklabels=[])
    ax2.set(yticklabels=[])
    plt.show()
```

```python
def comparision_visual_dist(original_score_list, final_score_list):
    ax1 = sns.kdeplot(original_score_list, shade=True, label='Association Score')
    ax2 = sns.kdeplot(final_score_list, shade=True, color='red', label='Calibrated Risk Score')
    ax1.set_xlim(0.0,1.0)
    ax2.set_xlim(0.0,1.0)
    ax1.set(yticklabels=[])
    ax2.set(yticklabels=[])
    plt.show()
```

```python
def version_hist(version_1_list, version_2_list):
    ax1 = sns.distplot(version_1_list, kde_kws={'label': 'version'})
    ax2 = sns.distplot(version_2_list, color='red', kde_kws={'label': 'version(updated)'})
    ax1.set_xlim(0.0,1.0)
    ax2.set_xlim(0.0,1.0)
    ax1.set(yticklabels=[])
    ax2.set(yticklabels=[])
    plt.show()
```
